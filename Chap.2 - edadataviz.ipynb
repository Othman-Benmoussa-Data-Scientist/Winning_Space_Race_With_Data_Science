## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
df.head(5)
```

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.


```python


# Scatter plot : Flight Number vs Launch Site
plt.figure(figsize=(10,6))
plt.scatter(df['FlightNumber'], df['LaunchSite'], alpha=0.6, c='blue')

plt.title("Flight Number vs Launch Site")
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['PayloadMass'], df['LaunchSite'], alpha=0.6, c='green')

plt.title("Payload Mass vs Launch Site")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Launch Site")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python

# On considère Mission_Outcome : "Success" ou "Failure"
# Créer une colonne binaire (1 = succès, 0 = échec)
df['Mission_Success'] = df['Outcome'].apply(lambda x: 1 if 'True' in x else 0)

# Calculer le taux de succès par orbite
orbit_success = df.groupby('Orbit')['Mission_Success'].mean().sort_values(ascending=False)

# Bar chart
plt.figure(figsize=(10,6))
orbit_success.plot(kind='bar', color='seagreen', edgecolor='black')

plt.title("Taux de succès par type d'orbite")
plt.xlabel("Type d'orbite")
plt.ylabel("Taux de succès (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot : Flight Number vs Orbit
plt.figure(figsize=(12,6))
plt.scatter(df['FlightNumber'], df['Orbit'], alpha=0.6, c='purple')

plt.title("Flight Number vs Orbit Type")
plt.xlabel("Flight Number")
plt.ylabel("Orbit Type")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python

plt.figure(figsize=(12,6))
plt.scatter(df['PayloadMass'], df['Orbit'], alpha=0.6, c='darkorange')

plt.title("Payload Mass vs Orbit Type")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Orbit Type")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Ajouter une colonne succès/échec (1 si succès, 0 sinon)
df['Mission_Success'] = df['Outcome'].apply(lambda x: 1 if 'True' in x else 0)

# Extraire l'année à partir de la date
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Calculer le taux de succès moyen par année
yearly_success = df.groupby('Year')['Mission_Success'].mean()

# Line chart
plt.figure(figsize=(10,6))
plt.plot(yearly_success.index, yearly_success.values, marker='o', linestyle='-', color='blue')

plt.title("Taux de succès annuel moyen des lancements SpaceX")
plt.xlabel("Année")
plt.ylabel("Taux de succès (%)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column

# Group by 'Orbit' and calculate the mean of 'Class' (i.e. success rate)
success_rate_by_orbit = df.groupby('Orbit')['Class'].mean().reset_index()

# Rename for clarity
success_rate_by_orbit.rename(columns={'Class': 'SuccessRate'}, inplace=True)

# Plot as bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=success_rate_by_orbit, x='Orbit', y='SuccessRate')

# Styling
plt.title('Success Rate by Orbit Type')
plt.ylabel('Success Rate')
plt.xlabel('Orbit')
plt.ylim(0, 1)  # Since it's a percentage (0 to 1)
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python


# Supposons que ton DataFrame final s'appelle df
# et qu'il contient les colonnes : 'LaunchSite' et 'LandingSuccess' (0/1)

# Calcul du taux de succès par site
success_rate = df.groupby('LaunchSite')['Class'].mean().reset_index()

# Bar chart
plt.figure(figsize=(8,6))
sns.barplot(x='LaunchSite', y='Class', data=success_rate, palette='Blues_d')

# Ajout des titres et labels
plt.title('Success Rate by Launch Site', fontsize=14)
plt.ylabel('Success Rate')
plt.xlabel('Launch Site')
plt.ylim(0,1)  # pour un taux entre 0 et 100%
plt.xticks(rotation=45)

plt.show()
```

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate

# Convertir 'Date' en entier (année)
df['Date'] = df['Date'].astype(int)

# Grouper par année et calculer la moyenne de 'Class' (taux de succès)
success_rate_by_year = df.groupby('Date')['Class'].mean().reset_index()

# Tracer le graphique
plt.figure(figsize=(10,6))
sns.lineplot(data=success_rate_by_year, x='Date', y='Class', marker='o')

plt.title('Taux de succès par année')
plt.xlabel('Année')
plt.ylabel('Taux de succès')
plt.ylim(0, 1)  # échelle de 0 à 1 pour un taux
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020


## Features Engineering


By now, you should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the future module.


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# HINT: Use get_dummies() function on the categorical columns

# Assuming 'features' is your dataframe containing all features including categorical columns

# List of categorical columns to encode
categorical_cols = ['Orbit', 'LaunchSite', 'LandingPad', 'Serial']

# Apply get_dummies() on these columns, drop first=False to keep all categories (optional)
one_hot_encoded = pd.get_dummies(features[categorical_cols], drop_first=False)

# Drop the original categorical columns from features
features_dropped = features.drop(columns=categorical_cols)

# Concatenate the one-hot encoded columns back to the dataframe
features_one_hot = pd.concat([features_dropped, one_hot_encoded], axis=1)

# Display the first few rows
print(features_one_hot.head())
```

### TASK  8: Cast all numeric columns to `float64`


Now that our <code>features_one_hot</code> dataframe only contains numbers, cast the entire dataframe to variable type <code>float64</code>


```python
# HINT: use astype function
features_one_hot = features_one_hot.astype('float64')
```
